```python
import utils
import matplotlib.pyplot as plt
```

```python
gru_loss = './eval_data/network_loss_gru.txt' 
lstm_loss = './eval_data/network_loss_lstm.txt'
gru_acc = './eval_data/network_accuracy_gru.txt'
lstm_acc = './eval_data/network_accuracy_lstm.txt'
losses = utils.graph_model_loss(gru_loss, lstm_loss)
acc = utils.graph_model_acc(gru_acc, lstm_acc)
```

```python
plt.xlabel('Iterations')
plt.ylabel('CrossEntropy Loss')
plt.title('Learning Curve Evaluation on MINST for LSTM and GRU')
plt.plot(losses.T[:,0], label='GRU Loss')
plt.plot(losses.T[:,1], label='LSTM Loss')
plt.legend()
plt.savefig('Learning Curve Evaluation on MINST for LSTM and GRU.png')
```

```python
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Learning Curve Evaluation on MINST for LSTM and GRU')
plt.plot(acc.T[:,0], label='GRU acc')
plt.plot(acc.T[:,1], label='LSTM acc')
plt.legend()
plt.savefig('Accuracy Curve Evaluation on MINST for LSTM and GRU.png')
```
